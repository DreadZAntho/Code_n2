```python
import pandas as pd

gender_submission=pd.read_csv('gender_submission.csv')
data_train=pd.read_csv('train.csv')
data_test=pd.read_csv('test.csv')

print(gender_submission)
print()
print(data_train)
print()
print(data_test)
```

```python
count_nan_key=data_train.isna().sum()
print(count_nan_key)
print()
print("There is : " + str((count_nan_key.sum()/data_train.size)*100) + " % of missing values in the train set")
```

```python
data_train['Age'].fillna(data_train['Age'].median(), inplace=True)
print(data_train.isna().sum())
```

```python
count_nan_key_test=data_test.isna().sum()
print(count_nan_key_test)
print()
print("There is : " + str((count_nan_key_test.sum()/data_test.size)*100) + " % of missing values in the test set")
```

```python
data_test['Age'].fillna(data_test['Age'].median(), inplace=True)
data_test['Fare'].fillna(data_test['Fare'].median(), inplace=True)
print(data_test.isna().sum())
```

```python
gender_submission = gender_submission.drop_duplicates()
print(gender_submission)
print()

data_train = data_train.drop_duplicates()
print(data_train)
print()

data_test= data_test.drop_duplicates()
print(data_test)
```

```python
print(gender_submission.describe())
print()

print(data_train.describe())
print()

print(data_test.describe())
```

```python
data_train["size_familly"] = data_train["Parch"] + data_train["SibSp"] + 1

sns.histplot(x="size_familly", hue="Survived", data=data_train)
plt.show()
```

```python
newDataTrain = data_train.drop(["Name", "PassengerId","Ticket","Cabin","Embarked","size_familly"],axis=1)
newDataTest = data_test.drop(["Name", "PassengerId","Ticket","Cabin","Embarked"],axis=1)

newDataTrain['Sex'] = newDataTrain['Sex'].replace(['male', 'female'], [1, 0])
newDataTest['Sex'] = newDataTest['Sex'].replace(['male', 'female'], [1, 0])
        
print(newDataTrain)
print()
print(newDataTest)
```

```python
y_train= newDataTrain["Survived"]
X_train = newDataTrain.drop("Survived",axis=1)
print(y_train)
print()
print(X_train)
print()

y_test = gender_submission["Survived"]
X_test = newDataTest
print(y_test)
print()
print(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

model=LogisticRegression()

param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'max_iter': [100, 500, 1000, 5000]
}

grid=GridSearchCV(model,param_grid,cv=5)

grid.fit(X_train,y_train)
print(grid.best_score_)
print(grid.best_params_)
print()

model = grid.best_estimator_
print(model.score(X_test,y_test))
```

```python
import numpy as np

def Survived(model,Pclass,Sex,Age,SibSp,Parch,Fare):
    x=np.array([Pclass,Sex,Age,SibSp,Parch,Fare]).reshape(1,6)
    print(model.predict(x))
    print(model.predict_proba(x))

Survived(model,2,1,18,1,0,30)
```
